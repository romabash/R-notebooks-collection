# Providing References to Plots

```r
library(readr)
library(dplyr)
library(ggplot2)
library(ggthemes)
```

### Load "worldcup" Dataset

```r
worldcup <- read_csv("Data/worldcup.csv")

worldcup <- worldcup %>%
  select(player_name, everything()) %>%
  rename(Name = player_name)

head(worldcup)
```

### Create a Scatterplot of Shots vs Passes for Forward Position

```r
# Filter rows for Forward Position
foward <- worldcup %>%
  filter(Position == 'Forward')

# Plot simple Scatterplot
ggplot(data = foward, aes(x = Passes, y = Shots)) +
  geom_point()
```

### Add References to improve the Plot:
- using theme-few() from "ggthemes" package
- Adding a smooth line using geom_smooth() function

```r
ggplot(data = foward, aes(x = Passes, y = Shots)) +
  geom_point() +
  theme_few() +
  geom_smooth()
```

### - geom_smooth() arguments:
- smoothing method (function) to use, eg. "lm", "glm", "gam", "loess", "rlm"
- Default method = "auto" the smoothing method is chosen based on the size of the largest group (across all panels). loess is used for less than 1,000 observations
- span argument controls how smooth the line will be.  Default  span = 0.75.  Higher means smoother

```r
ggplot(data = foward, aes(x = Passes, y = Shots)) +
  geom_point() +
  theme_few() +
  geom_smooth(method = 'lm')
```

#### level argument sets the confidence interval level: Default is 95%

```r
ggplot(data = foward, aes(x = Passes, y = Shots)) +
  geom_point() +
  theme_few() +
  geom_smooth(method = 'lm', level = 0.90)
```

#### se argument indicates whether to include confidence intervals: Default is TRUE

```r
ggplot(data = foward, aes(x = Passes, y = Shots)) +
  geom_point() +
  theme_few() +
  geom_smooth(method = 'lm', se = FALSE)
```
