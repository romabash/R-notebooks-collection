# Customizing ggplot2 Plots 
- Using dplyr for data cleaning
- Using ggplot2, gridExtra and ggthemes for plotting

```r
library(readr)
library(dplyr)
library(ggplot2)
library(gridExtra)
library(ggthemes)
```

# Using themes

### Load "worldcup" Dataset

```r
worldcup <- read_csv("Data/worldcup.csv")

worldcup <- worldcup %>%
  select(player_name, everything()) %>%
  rename(Name = player_name)

head(worldcup)
```

### Create a plot:

```r
worldcup_plot <- ggplot(data = worldcup, aes(x = Time, y = Shots)) +
  ggtitle("Shots vs Time") +
  xlab("Time") +
  ylab("Shots")
```

### theme_gray()
- The signature ggplot2 theme with a grey background and white gridlines, designed to put the data forward yet make comparisons easy

```r
worldcup_plot + 
  geom_point() +
  theme_gray()
```

### theme_bw()
- The classic dark-on-light ggplot2 theme. May work better for presentations displayed with a projector

```r
worldcup_plot + 
  geom_point() +
  theme_bw()
```

### theme_linedraw()
- A theme with only black lines of various widths on white backgrounds, reminiscent of a line drawings. Serves a purpose similar to theme_bw
- Lines look lighter in Jupyter Notebook than in RStudio

```r
worldcup_plot + 
  geom_point() +
  theme_linedraw()
```

### theme_light()
- A theme similar to theme_linedraw but with light grey lines and axes, to direct more attention towards the data
- Similar to theme_bw() but without a dark border

```r
worldcup_plot + 
  geom_point() +
  theme_light()
```

### theme_dark()
- Similar to theme_light, with similar line sizes but a dark background. Useful to make thin coloured lines pop out

```r
worldcup_plot + 
  geom_point() +
  theme_dark()
```

### theme_minimal
- A minimalistic theme with no background annotations
- Similar to theme_light() without a light border

```r
worldcup_plot + 
  geom_point() +
  theme_minimal()
```

### theme_classic()
- A classic-looking theme, with x and y axis lines and no gridlines

```r
worldcup_plot + 
  geom_point() +
  theme_classic()
```

### theme_void()
- A completely empty theme

```r
worldcup_plot + 
  geom_point() +
  theme_void()
```

### theme_solarized()
- From "ggthemes" package

```r
worldcup_plot + 
  geom_point() +
  theme_solarized()
```

###  theme_few() 
- From "ggthemes" package

```r
worldcup_plot + 
  geom_point() +
  theme_few()
```
