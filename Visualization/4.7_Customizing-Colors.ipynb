# Colors and Fill Aesthetics
- For categorical variables (gender), use scale_color_hue() or scale_color_manual()
- For continuous variable (age), use scale_color_gradient()
- Custom scale functions with "RColorBrewer" package, using "Brewer" color palettes
  - Pick a Brewer palette and specify it with palette argument within the scale_color_brewer() function

```r
library(readr)
library(dplyr)
library(ggplot2)
library(gridExtra)
library(ggthemes)

library(RColorBrewer)
```

#### Load Data

```r
worldcup <- read_csv("Data/worldcup.csv")

worldcup <- worldcup %>%
  select(player_name, everything()) %>%
  rename(Name = player_name)

head(worldcup)
```

## Display palettes:

```r
display.brewer.pal(name="Set1", n = 8)
```

```r
display.brewer.pal(name="PRGn", n = 8)
```

```r
display.brewer.pal(name="PuBuGn", n = 8)
```

### Example: World Cup 2010 data
- Plot of Time vs Passes for the "worldcup" dataset
- Number of Shots taken is represented by Size
- Position is represented by Color

```r
wc_data <- worldcup %>%
  ggplot() +
  theme_bw() +
  geom_point(aes(x = Time, y = Passes, color = Position, size = Shots), alpha = 0.5)

wc_data
```

#### Set1 palette

```r
a <- wc_data +
  scale_color_brewer(palette = "Set1") +
  ggtitle("Set1 Palette")
a
```

#### Dark2 palette

```r
b <- wc_data +
  scale_color_brewer(palette = "Dark2") +
  ggtitle("Dark2 Palette")
b
```

#### Accent palette

```r
c <- wc_data +
  scale_color_brewer(palette = "Accent") +
  ggtitle("Accent Palette")
c
```

#### Pastel2 palette

```r
d <- wc_data +
  scale_color_brewer(palette = "Pastel2") +
  ggtitle("Pastel2 Palette") +
  theme_dark()
d
```

### Use grid.arrange from "gridExtra" package to arrange all Plots on one Graph

```r
grid.arrange(a, b, c, d, ncol = 2)
```

### Set Discrete colors:
- Set manually using scale_color_manual() and scale_fill_manual() functions
- Lines and Shape points take only "color" aesthetic, Use fill for other Shapes like polygons...
- Provide vector of colors with the same number of arguments 

```r
worldcup %>%
  ggplot() +
  theme_bw() +
  geom_point(aes(x = Time, y = Passes, color = Position, size = Shots), alpha = 0.5) +
  scale_color_manual(values = c("blue", "red", "darkgreen", "darkgray"))
```
