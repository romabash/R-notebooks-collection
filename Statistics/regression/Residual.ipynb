# Residual Plot
- The residual data of the simple linear regression model is the difference between the observed data of the dependent variable and the fitted values
    - e = Yobserved - Ypredicted
    - Observed is the actual point on the scatter plot
    - Predicted is the Regression line
- A residual is the vertical distance between a data point and the regression line
- Each data point has one residual
- They are positive if they are above the regression line and negative if they are below the regression line
- If the regression line actually passes through the point, the residual at that point is zero
- As residuals are the difference between any data point and the regression line, they are sometimes called “errors” 

```r
head(faithful)
```

```r
eruption.lm = lm(eruptions ~ waiting, data=faithful) 
```

#### Compute the residual with the resid() function

```r
eruption.res = resid(eruption.lm)
head(eruption.res)
```

#### Plot the residual against the observed values of the variable waiting

```r
plot(faithful$waiting, eruption.res, ylab="Residuals", xlab="Waiting Time", main="Old Faithful Eruptions") 
abline(0, 0)  # the horizon
```

# The standardized residual is the residual divided by its standard deviation

```r
eruption.lm = lm(eruptions ~ waiting, data=faithful) 
```

#### Compute the standardized residual with the rstandard function

```r
eruption.stdres = rstandard(eruption.lm)
head(eruption.stdres)
```

#### Plot the residual against the observed values of the variable waiting

```r
plot(faithful$waiting, eruption.stdres, ylab="Residuals", xlab="Waiting Time", main="Old Faithful Eruptions") 
abline(0, 0)  # the horizon
```

### Normal Probability Plot of Residuals

```r
qqnorm(eruption.stdres, ylab="Standardized Residuals", xlab="Normal Scores", main="Old Faithful Eruptions") 
qqline(eruption.stdres)
```
