# Rank (Spearman) and Partial Correlation
- Pearson correlation measures Linear relationship
    - When one variable increases then other variable increases by consistent amount
- Spearman correlation measures non-linear relationship
    - When one variable increases then other variable increases by non-consistent amount
- Spearman correlation is often used to evaluate relationships involving ordinal variables
    - Used often in Sports (How well judges agree with each other)
- Spearman Correlation
    - Correlation Ranks: from Lowest to Highest (or highest to lowest)
    - If the same number, assign its average
    - r = 1 – ( 6*Sum(Rank Difference)2 / n(n2 – 1))

|    X    |    Y    |  Rank X | Rank Y  |  X – Y  | (X-Y)^2 |

## Spearman vs Pearson Correlation

```r
clubHeadSpeed = c(100, 102, 103, 101, 105, 100, 99, 105)
distance = c(257, 264, 274, 266, 277, 263, 258, 275)
x1 = clubHeadSpeed
y1 = distance
cor(x1,y1, method="spearman")
```

#### Compared to Pearson

```r
cor(x1,y1, method="pearson")
```

## Examples:

#### Example 1: Both at +1

```r
x1 = seq (1,10,0.1)
y1 = x1
plot(x1,y1)
cor(x1,y1, method="pearson")
 
cor(x1,y1, method="spearman")
```

#### Example 2: Both increase but not Linear.  Spearman ranks are still increasing ans still remains 1

```r
x1 = seq (1,10,0.1)
y1 = x1^2
plot(x1,y1)
cor(x1,y1, method="pearson")
 
cor(x1,y1, method="spearman")
```

#### Example 3 Both increase but not linear.  Spearman ranks are still increasing

```r
x1 = seq (-10,20,0.1)
y1 = 0.25*(x1-4)^3 + 3
plot(x1,y1)
cor(x1,y1, method="pearson")
 
cor(x1,y1, method="spearman")
```

#### Example 4 – Negative correlation

```r
x1 = seq (1,10,0.1)
y1 = -x1
plot(x1,y1)
cor(x1,y1, method="pearson")
 
cor(x1,y1, method="spearman")
```

#### Example 5 – Non-Linear Negative correlation. Spearman ranks are still decreasing

```r
x1 = seq (-10,10,0.1)
y1 = -0.25*x1^3 + 3
plot(x1,y1)
cor(x1,y1, method="pearson")
 
cor(x1,y1, method="spearman")
```

#### Example 6: Zero correlation in both cases

```r
x1 = seq (-10,10,0.1)
y1 = x1^2
plot(x1,y1)
cor(x1,y1, method="pearson")
 
cor(x1,y1, method="spearman")
```

# Partial Correlation
- If 2 variables are correlated
    - Cannot conclude that they have casual relationship
    - Third variable that explains the relationship

### Example:
- Correlation between Shoe Size and Reading Ability:
- Third Variable: Age
- Look at Correlation between:
    - Shoe Size and Reading Ability: r=0.80
    - Shoe Size and Age: r=0.90
    - Reading Ability and Age: r=0.85

### Partial Correlation:
- Correlation between 2 Variables without the third
    - Correlation between Show Size and Reading Ability with Age removed
    - rxy.z = rxy – (rxz*ryz) / Sqrt( (1 – rxz2)(1 – ryz2) )
    - rxy.z = 0.80 – (0.90 * 0.85) / Sqrt( (1 – 0.902)(1 – 0.852) ) = 0.152
- Correlation between Show Size and Reading Ability went from 0.80 to 0.152 when Age Variable is removed
