## Normal Distribution (Empirical Rule)
- In a normal distribution (Bell Shape Curve), nearly all of the data will fall within three standard deviations of the mean. The empirical rule can be broken down into three parts:
- 68% of data falls within the first standard deviation from the mean (Mean +/- 1SD)
- 95% fall within two standard deviations (Mean +/- 2SD)
- 99.7% fall within three standard deviations (Mean +/- 3SD)
- 34% + 13.5% + 2.35% + 0.15% = 50%

#### Example: Student IQ Test
- Mean = 100
- Standard Deviation = 15
- 68% of the data will be within 85 - 115
- 100 – 15 = 85
- 100 + 15 = 115

### Normal Distribution
- SD changes how narrow or wide the curve is (Smaller Sigma: Narrow Curve, Higher Sigma: Wider Curve)
- Change in Mean, shifts the Curve (Higher Mean shifts Curve to the right, Lower Mean shifts Curve to the left)
- Symmetric about its mean 𝜇
- Mean = Median = Mode
    - Single peak at x = Mean
- Inflection point at (Mean - 1SD) and (Mean + 1SD), where slope starts to change
- Area under the curve = 1
- Area of left Mean = Area of right of Mean = ½

### Standard Normal Distribution
- Mean = 0
- SD = 1
- Mean = Median = Mode
    - Single peak at z = 0
- Inflection point at -1 and 1, where slope starts to change

#### Find the z-value based on the Mean and SD:

```r
m <- 100
s <- 15
d <- 110
zValue <- (d - m)/s
zValue
```

#### Compute ‘Left’ Area (Probability) under the Standard Normal Curve given the z-value:

```r
z <- 0.21
pnorm(z)
```

#### Compute ‘Right’ Area (Probability) under the Standard Normal Curve given the z-value:

```r
z <- 0.21
1 - pnorm(z)
```

#### Compute ‘In Between’ Area (Probability) under the Standard Normal Curve given 2 z-values:

```r
z1 = -0.31
z2 = 0.31
pnorm(z2) - pnorm(z1)
```

#### Given ‘Left’ Area (Probability) under the Standard Normal Curve = 0.58, Compute z value

```r
p <- 0.58
zValue <- qnorm(p)
zValue
```

## Testing Normality
- QQ plot – Quantile-Quantile plot
- Analytical/Graphical technique: Data is plotted against a theoretical normal distribution
    - Straight line means Data is normally distributed
- First Sort the data
- Plot against appropriate quantiles from the standard normal distribution

#### Example:
- Divide the normal distribution curve into (n+1) parts
- Each part represents % of the area
    - For sample size of 9, each part represents 10% (n+1)
- Compute the corresponding z-values of each segment: i/(n+1)
    - 10% = -1.28
    - 20% = -0.84
    - 30% = -0.52
    - 40% = -0.25
    - 50% = 0
    - 60% = 0.25
    - 70% = 0.52
    - 80% = 0.84
    - 90% = 1.28
    - 100% = 3.0
- QQ plot
    - X axis: z-values taken from the standard normal distribution curve
    - Y-axis: Sorted Data values

```r
x <- c(3.89,4.75,6.33,4.75,7.21,5.78,5.80,5.20,6.64)
qqnorm(x)
qqline(x)
```

#### Example: Known Normal Distribution

```r
x <- rnorm(n=10,mean=5,sd=2)
hist(x)
 
#qqplot:
qqnorm(x)
qqline(x)
```

#### Example: With n=1000 for better visualization

```r
x <- rnorm(n=1000,mean=5,sd=2)
hist(x)
 
#qqplot:
qqnorm(x)
qqline(x)
 
```

### Example: Known Uniform Distribution

```r
x <- runif(n=1000)
hist(x)
 
#qqplot:
qqnorm(x)
qqline(x)
```
