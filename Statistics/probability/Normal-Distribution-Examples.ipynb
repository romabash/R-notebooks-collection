# Normal Distribution 

### dnorm() Function
- This function gives height of the probability distribution at each point for a given mean and standard deviation.

```r
# Create a sequence of numbers between -3 and 3 incrementing by 0.1.
x <- seq(-3, 3, by = .1)
# Choose the mean as 2.5 and standard deviation as 0.5.
y <- dnorm(x, mean = 0, sd = 1)

# Give the chart file a name.
#png(file = "dnorm.png")

plot(x,y)

# Save the file.
#dev.off()
```

### pnorm() Function
- Gives the probability of a normally distributed random number to be less that the value of a given number. It is also called "Cumulative Distribution Function"

```r
# Create a sequence of numbers between -3 and 3 incrementing by 0.1.
x <- seq(-3, 3, by = .1)
# Choose the mean as 2.5 and standard deviation as 0.5.
y <- pnorm(x, mean = 0, sd = 1)

plot(x,y)
```

### qnorm()
- Takes the probability value and gives a number whose cumulative value matches the probability value

```r
# Create a sequence of probability values incrementing by 0.02
x <- seq(0, 1, by = 0.02)
# Choose the mean as 2 and standard deviation as 3
y <- qnorm(x, mean = 2, sd = 3)

plot(x,y)
```

### rnorm()
- Used to generate random numbers whose distribution is normal. It takes the sample size as input and generates that many random numbers. We draw a histogram to show the distribution of the generated numbers.

```r
# Create a sample of 50 numbers which are normally distributed.
y <- rnorm(50)

# Plot the histogram for this sample.
hist(y, main = "Normal DIstribution")
```
