# Reading Data

- read.table, read.csv to read tabular data (inverse of write.table)
- readLines, to read lines of text file (inverse of writeLines)
- source, to read in R code files (inverse of dump)
- dget, to read in R code files (invesre of dput)
- unserialize, to read single R object in binary form (inverse of serialize)

## read.table() function arguments:
- file, name of file
- header, logical indicator if the file has a header line
- sep, indicate how columns are separated
- colClasses, character vector indicating class of each column
- nrows, number of rows
- comment.char, character string indicating comment character. Set to "", if there are no comment lines in the file
- skip, number of lines to skip
- stringsAsFactors, should character variables be coded as factors (default is TRUE)

## Reading in Larger Datasets
- use colClasses argument to specify classes of each column to make it run much faster
- read in a part of the file with nrows, then assign class to each column in the data 

```r
initial <- read.table("Data/team_standings.csv", sep=',', nrows=10, header=TRUE)
initial
```

```r
classes <- sapply(initial, class)
classes
```

```r
fullData <- read.table("Data/team_standings.csv", sep=',', header=TRUE, colClasses=classes)
fullData
```

### read.csv() function:
- identical to read.table() except that the default separator is a comma and header is defalut to TRUE

```r
data <- read.csv("Data/team_standings.csv")
data
```
