# Random Forest

```r
library(randomForest)
```

## Using *mtcars* dataset

```r
head(mtcars)
```

## Use Random Forest to predict mpg of a car based on varibales in the dataset
- Specify number of trees with *ntree* parameter to be 1000 
- Set *importance=TRUE* to look at the plot of important variables 

```r
set.seed(123)
rf <- randomForest(mpg ~ ., data = mtcars, ntree = 1000, importance = TRUE)
rf
```

## Plot predictors importance using the *varImpPlot* function

```r
varImpPlot(rf)
```

# Random Forest for Proximity Classification

## Measuring the proximity between records in dataset
- Using the **iris** dataset
- Classify based on *Sepal.Length*, *Sepal.Width*, *Petal.Length*, and *Petal.Width*

```r
head(iris)
```

```r
# Use the dataframe without the Species Column
# Provide proximity parameter proximity=True, and type="unsupervised"
irf <- randomForest(iris[, -5], proximity=TRUE, type="unsupervised")
irf
```

## Use the *multidimentional scaling plot* from *randomForest* package
- Pass in the **Species** as an argument to use as color

```r
MDSplot(irf, iris$Species)
```

## Plot the scatterplot of the acutaul data: Petal.Length, and Petal.Width
- Better separation than **randomForest** classification (overfitting)

```r
plot(iris[,3], iris[,4], col=iris[,5])
```
